```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV 파일 경로
csv_file = "total_rewards.csv"

# CSV 파일에서 데이터 읽어오기
data = pd.read_csv(csv_file)

# 그래프 그리기
plt.plot(data['Episode'], data['Total Reward'])

# 그래프 제목과 레이블 추가
plt.title('Reward per Episode')
plt.xlabel('Episode')
plt.ylabel('Reward')

# 그래프 표시
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV 파일 경로
csv_file = "total_rewards100_edit.csv"

# CSV 파일에서 데이터 읽어오기
data = pd.read_csv(csv_file)

# 에피소드 20까지의 데이터 선택
data = data[data['Episode'] <= 20]

# 그래프 그리기
plt.plot(data['Episode'], data['Total Reward'])

# 그래프 제목과 레이블 추가
plt.title('Reward per Episode')
plt.xlabel('Episode')
plt.ylabel('Reward')

# 그래프 표시
plt.show()
```
