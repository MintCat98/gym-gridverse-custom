```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV  
csv_file = "total_rewards100_without_closer.csv"

# CSV   
data = pd.read_csv(csv_file)

#  
plt.plot(data['Episode'], data['Total Reward'])

#    
plt.title('Reward per Episode')
plt.xlabel('Episode')
plt.ylabel('Reward')

#  
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

#  
data_100 = pd.read_csv("total_rewards100_without_closer.csv")
data_200 = pd.read_csv("total_rewards200_without.csv")
data_300 = pd.read_csv("total_rewards.csv")

#       
print(data_100.columns)
print(data_200.columns)
print(data_300.columns)
data_200['Episode'] += 100
data_300['Episode'] += 200

#  
df_vertical_concat = pd.concat([data_100, data_200, data_300], axis=0, ignore_index=True)

#   
print(df_vertical_concat.head())
print(df_vertical_concat.tail())

#  
plt.plot(df_vertical_concat['Episode'], df_vertical_concat['Total Reward'])

#    
plt.title('Reward per Episode')
plt.xlabel('Episode')
plt.ylabel('Reward')

#  
plt.show()
```
