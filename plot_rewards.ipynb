```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV  
csv_file = "total_rewards.csv"

# CSV   
data = pd.read_csv(csv_file)

#  
plt.plot(data['Episode'], data['Total Reward'])

#    
plt.title('Reward per Episode')
plt.xlabel('Episode')
plt.ylabel('Reward')

#  
plt.show()
```
